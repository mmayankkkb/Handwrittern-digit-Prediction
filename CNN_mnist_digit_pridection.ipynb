```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,MaxPool2D,Flatten,Conv2D,Dropout
```

```python
(X_train,Y_train),(X_test,y_test)=mnist.load_data();    # importing the dataset from keras
```

```python
X_test=X_test.reshape((X_test.shape[0],X_test.shape[1],X_test.shape[2],1))    
```

```python
X_train=X_train.reshape((X_train.shape[0]),X_train.shape[1],X_train.shape[2],1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#Normalizing the pixel values
X_train=X_train/255
X_test=X_test/255
print(X_test)
```

```python
print(Y_train)
print(y_test)
```

```python
model=Sequential()
```

```python
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))              #specifying 32 fillers of size (3*3)
                                                                                 #input size=28 by 28 and are greyscale(1)
```

```python
model.add(MaxPool2D(2,2))
```

```python
# Adding fully connected Layer
model.add(Flatten())
model.add(Dense(100,activation='relu'))
```

```python
#Output layer
model.add(Dense(10,activation='softmax'))
```

```python
#compilation
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=20)
```

```python
model.evaluate(X_test,y_test)
```
